```python
import os, sys
import glob
import pandas as pd
import numpy as np

def test():
    print("Test")
    
#class expertProcessor():

def processInputFiles():

    print("Hello")

    path = os.getcwd()+os.path.sep
    #dirs = glob.glob(path + 'Tristan Glatard - Evaluation_removed Comments.csv')
    listOfFiles = glob.glob(path +'inputFiles'+os.path.sep+ '*.csv')

    dataFrame = pd.DataFrame()
    counter = 0
    for inputFile in listOfFiles:
        counter += 1

        newDataFrame = pd.read_csv(inputFile, index_col=0)   #Set first column as index
        #newDataFrame.index = newDataFrame["Pipelines/Datasets"]   #Resplca index with first column

        newDataFrame = newDataFrame.replace(['x', 'X'],+1)
        newDataFrame = newDataFrame.replace([np.NaN],-1)
        newDataFrame.to_excel(inputFile.replace(".csv", ".xlsx")) 

        if dataFrame.empty:
            dataFrame = newDataFrame.copy()
            #dataFrame.to_excel(path +os.path.sep +'inputFiles'+os.path.sep+ "concatenatedDataFrames_" + str(counter) + ".xlsx")
            continue

        #Adding row and columns to dataFrame
        for rowName in newDataFrame.index:
            if not (rowName in dataFrame.index):
                #print("Found new Pipeline : ", rowName)
                dataFrame.loc[rowName] = 0

        for columnName in newDataFrame.columns:
            if not (columnName in dataFrame.columns):
                #print("Found new Dataset : ", columnName)
                dataFrame[columnName] = 0

        #Solving the conflicts
        for rowName in newDataFrame.index:  #Name of Pipelines
            for columnName in newDataFrame.columns:
                dataFrame.loc[rowName,columnName] += newDataFrame.loc[rowName,columnName]
                '''
                #print(rowName,columnName)
                if newDataFrame.loc[rowName,columnName] == 1:
                    dataFrame.loc[rowName,columnName] = 1
                elif newDataFrame.loc[rowName,columnName] == -1:
                    #if (np.isnan(dataFrame.loc[rowName,columnName])) or (dataFrame.loc[rowName,columnName] == 0):
                    if (dataFrame.loc[rowName,columnName] == 0):
                        dataFrame.loc[rowName,columnName] = -1
                #print("Conflict in ", os.path.basename(inputFile) ," @ ", dataFrame.loc[rowName,columnName], "=>", newDataFrame.loc[rowName,columnName])
                '''
    #Majority Voting to Normal
    for rowName in dataFrame.index:  #Name of Pipelines
        for columnName in dataFrame.columns:
            if (dataFrame.loc[rowName,columnName] > 0):
                dataFrame.loc[rowName,columnName] = 1
            elif (dataFrame.loc[rowName,columnName] < 0):
                dataFrame.loc[rowName,columnName] = -1


    dataFrame = dataFrame.sort_values(by=['Pipelines/Datasets'])
    #dataFrame = dataFrame.sort_values(by = 'Piplines/Datasets', axis = 'index')
    dataFrame = dataFrame.reindex(sorted(dataFrame.columns), axis=1)

    dataFrame.to_excel(os.path.join(os.getcwd(), 'inputFiles', "concatenatedDataFrames.xlsx")) 
    dataFrame.to_excel(os.path.join(os.getcwd(),"outputFile.xlsx")) 
    dataFrame.to_csv(os.path.join(os.getcwd(),"outputFile.csv"), sep=',', encoding='utf-8')
    print(dataFrame.shape)

    return 

if __name__ == "__main__":
    processInputFiles()
    
    
    
    
```
